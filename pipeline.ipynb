```python
qa=pipeline("question-answering")
```

```python
c='''
Pawan Kalyan, widely known by his initials PSPK,
is a prominent Indian actor, politician, and
 philanthropist who primarily works in the
 Telugu film industry. Renowned for his
 charismatic screen presence and unique
 style, he gained immense popularity with
  blockbuster films like Tholi Prema,
  Gabbar Singh,and Attarintiki Daredi. Beyond cinema,
   Pawan Kalyan is deeply committed to public service.
    He founded the JanaSena Party in 2014, aiming to
     bring change through clean and people-centric
     politics in Andhra Pradesh and Telangana.
      Known for his simplicity, discipline,
      and philosophical outlook,
      Pawan Kalyan is not only a beloved actor but also a passionate
leader who continues to inspire millions through both his films and his political ideals.
'''
```

```python
q="in which year he formed janasena party ?"
```

```python
result=qa({'question':q,'context':c})
print(result)
```
