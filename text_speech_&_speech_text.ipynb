```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
ip='/content/WhatsApp Image 2025-05-20 at 11.12.14 AM.jpeg'
plt.imshow(Image.open(ip))
plt.axis('off')
```

```python
img = Image.open(ip)
text=pytesseract.image_to_string(img)
print("Extended text",text)
```
